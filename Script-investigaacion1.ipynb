```python
# Declaramos el diccionario
def crear_data(x):
    tabla = dict()    
    if type(x)==list:
        
        for i in range(len(x)):
    
            tabla['tweets_list'+str(i+1)] = []
    else:
       for i in range(len(x)):
            tabla['tweets_df'+str(i+1)]  = pd.DataFrame(x['tweets_list'+str(i+1)], columns=['Datetime', 'Tweet Id', 'Text', 'Username'])         
    return(tabla)
```
